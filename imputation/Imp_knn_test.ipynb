```python
# Import necessary libraries
import pandas as pd
from sklearn.impute import KNNImputer
import numpy as np
import os
```

```python
# change current working directory to a new directory
os.chdir('/rds/general/project/hda-22-23/live/TDS/Group5/Models')

# check if the current working directory has been changed
print(os.getcwd())

test_df = pd.read_csv('test_df.csv')
print(test_df)
```

```python
# Get the number of columns in the test dataframe
len(test_df.columns)

# Get the length of the eid column
len(test_df['eid'])

# Get the number of unique values in the eid column
len(test_df['eid'].unique())
```

```python
# cleaning data - we want to drop health_score and unnamed column
test_df = test_df.drop('health_score', axis=1)
```

```python
# making all the values that are less than 0 to NA in mood swing, smoking and drinking for imputation
test_df['Smoking_status.0.0'][test_df['Smoking_status.0.0'] < 0] = np.nan
test_df['Mood_swings.0.0'][test_df['Mood_swings.0.0'] < 0] = np.nan
test_df['Alc_drinker_status.0.0'][test_df['Alc_drinker_status.0.0'] < 0] = np.nan
```

```python
missing_count = test_df['Mood_swings.0.0'].isnull().sum()

print("Number of missing values in column Mood_swings.0.0:", missing_count)
```

```python
# get all column names
col_names = test_df.columns

print(col_names)
```

```python
subset_test_df = test_df[['Age_recr.0.0', 'Sex.0.0', 'BMI.0.0', 'Mood_swings.0.0', 'Smoking_status.0.0', 'Alc_drinker_status.0.0', 'Lymphocyte_count.0.0', 'Monocyte_count.0.0', 'Reti_count.0.0', 'WBC_count.0.0', 'RBC_count.0.0', 'Hgb_conc.0.0', 'Haematocrit_perc.0.0', 'Platelet_count.0.0', 'Basophil_count.0.0', 'Eosinophil_count.0.0', 'Neutrophil_count.0.0', 'Immature_ret_fraction.0.0', 'High_light_scatter_reti_count.0.0', 'C_reactive_protein.0.0', 'Traff_int_major_rd.0.0', 'housing_score']]
print(subset_test_df)
```

```python
# Initialize KNNImputer object with desired arguments
imputer = KNNImputer(n_neighbors=5)

# Impute missing values
imputed_data = imputer.fit_transform(subset_test_df)
```

```python
imputed_df = pd.DataFrame(imputed_data, columns=subset_test_df.columns)
```

```python
# Check for missing values in imputed_df and rounding up decimal numbers from ordinal variables
print(imputed_df.isna().sum())
imputed_df['Mood_swings.0.0'] = imputed_df['Mood_swings.0.0'].apply(lambda x: round(x))
imputed_df['Smoking_status.0.0'] = imputed_df['Smoking_status.0.0'].apply(lambda x: round(x))
imputed_df['Alc_drinker_status.0.0'] = imputed_df['Alc_drinker_status.0.0'].apply(lambda x: round(x))
```
